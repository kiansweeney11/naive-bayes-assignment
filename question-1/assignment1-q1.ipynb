```python
# load in libraries
import math
import re
import os
import sys
```

Three step approach:
- read in file
- process text
- split into sentence

```python
def load_text_file(file):
    # had unicode error without encoding, most text files are utf-8 encoded anyway
    with open(file, encoding="utf8") as f:
        # open file then create string of all the lines
        listToStr = ' '.join([str(elem) for elem in f])
        # remove newline characters
        str2 = listToStr.replace("\n", "")
        return str2
```

```python
load_text_file('test.txt')
```

```python
### Rules
# multiple ... in sentence -> not end of sentence
# .  followed by capital -> not necessarily a sentence (Mr. Tyler Perry)
# !, ? followed by capital -> sentence?
# words like Mr., Mrs., Dr., St., Ave. will all need handling as not sentence
```

## Attempt 1

```python
def split_into_sentence(file):
    sent = []
    text = load_text_file(file)
    # there is a detailed breakdown of this regular expression at the end of this file
    m = re.split(r'(?<=[^A-Z].[.?!]) (?=[A-Z])', text)
    for i in m:
        sent.append(i)
    str1 = " " 
    return sent
```

```python
split_into_sentence('test.txt')
```

##### It works for our test case. Let's try some unseen sentences.

```python
with open('test-1-output.txt', 'w', encoding="utf8") as f:
    f1 = split_into_sentence('test.txt')
    for item in range(len(f1)):
        f.write(f1[item] + "\n")
```

## Constraints

It seems our regex expression cannot handle capitialized letters after a punctuation such as "R." like here for example. When I break down our expression it makes sense why it does not work. Here's why:
- The first part of the expression uses negation to find a non capitalized letter. ([^A-Z])
- The next part locates the typical end of sentence punctuations used in English. ([.?!])
- Lastly, it optionally locates a capitialized letter at the start of the next word.

The first part of this if ultimately why this expression fails as it fails to deal with a capital at start of string like "R." or the U.S. abbrevation before a capital also. I asked on forums online to see if capturing this case is viable for regex and the consensus I got was that it was not.
